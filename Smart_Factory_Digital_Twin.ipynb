**Downloading the Dataset**

```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("zoya77/decentralized-smart-factory-operational-data")
arquivos = os.listdir(path)
csv_files = [f for f in arquivos if f.endswith('.csv')]

arquivo_correto = os.path.join(path, csv_files[0])
df = pd.read_csv(arquivo_correto)
df.head()
```

**Getting info from the dataset**

```python
df.info()
```

**Data Preprocessing**

**1. Feature Selection**

We can remove the Machine ID and Hour columns, because it doesn't matter for the physics analysis.





```python
target_col = 'Production_Rate'

drop_cols = [target_col, 'Machine_ID', 'Hour']
X = df.drop(columns=[c for c in drop_cols if c in df.columns])
y = df['Production_Rate']

drop_cols = [target_col, 'Machine_ID','Hour']
X = df.drop(columns=[c for c in drop_cols if c in df.columns])
y = df['Production_Rate']
```

**2. Handling Missing Values**

We know that this current dataset have no missing values, but for a real life problem it might have some missing values that we need to clean to keep the production system stability.

```python
from sklearn.impute import SimpleImputer
if X.isna().sum().sum() > 0:
  imputer = SimpleImputer(strategy = 'mean')
  X_imputed = pd.DataFrame(imputer.fit_transform(X), columns = X.columns)
else:
  print(f"Dataset is clean!")
  X_imputed = X.copy()
```

**3. Scaling**

Probably, we're gonna have to compare 2 or more models to see which one has the better accuracy for our problem. So, we're gonna normalize the data to ensure a fair comparison

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler()

X_train_raw, X_test_raw, y_train, y_test = train_test_split(X_imputed, y, test_size = 0.2, random_state = 42)

X_train = pd.DataFrame(scaler.fit_transform(X_train_raw), columns = X.columns)
X_test = pd.DataFrame(scaler.transform(X_test_raw), columns = X.columns)
```

**Now it's time to compare the models**

image.png

Following the path in the image, the best models to use are either: **RidgeRegression**,**SVR** or **XGBoost (one of the Ensemble Regressors)**. Let's compare their MAE and RÂ² Score.

```python
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
import xgboost as xgb

models = {
    "Ridge Regression": Ridge (alpha = 1.0),
    "SVR (RBF Kernel)": SVR(C=1.0, epsilon = 0.1),
    "XGBoost (Ensemble Regressors)": xgb.XGBRegressor(objective='reg:squarederror',n_estimators = 100, random_state = 42)
}

results = {}

for name, model in models.items():
  model.fit(X_train,y_train)
  preds = model.predict(X_test)

  mae = mean_absolute_error(y_test, preds)
  r2 = r2_score(y_test, preds)

  results[name]= mae

  print (f"Model: {name}")
  print(f"   -> Mean Absolute Error (MAE): {mae:.4f}")
  print(f"   -> RÂ² Score: {r2:.4f}")
  print("="*30)

best_model = min(results, key = results.get)
print(f"Best Model: {best_model}")
```

**The Ridge Regression model fits the best for our problem, let's check which coefficient is positive or negative for the Production Rate**

```python
import matplotlib.pyplot as plt
import seaborn as sns

winner_model = models["Ridge Regression"]

coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Weight': winner_model.coef_
})

coefficients ['Absolute Importance'] = coefficients['Weight'].abs()
coefficients = coefficients.sort_values(by = 'Absolute Importance', ascending = False)

plt.figure(figsize=(10, 6))

sns.barplot(x='Weight', y='Feature', data=coefficients, palette='coolwarm')

plt.title('Sensitivity Analysis: What drives Production Rate?', fontsize=16)
plt.xlabel('Impact Magnitude (Left = Reduces Production | Right = Increases Production)', fontsize=12)
plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

top_feature = coefficients.iloc[0]['Feature']
top_weight = coefficients.iloc[0]['Weight']
direction = "INCREASES" if top_weight > 0 else "DECREASES"

print(f"\n>>> KEY FINDING:")
print(f"The most critical variable is '{top_feature}'.")
print(f"Mathematical Insight: For every unit increase in {top_feature}, the Production Rate {direction} by {abs(top_weight):.2f} units.")
```

The sensitivity analysis reveals that the factory operates as a demand-driven system, with ***Demand_Response*** being the strongest positive driver (+30.17) of the Production Rate. Conversely, ***Power_Consumption*** acts as the primary bottleneck (-15.0), indicating that energy inefficiencies or peak load conditions significantly penalize throughput. Surprisingly, control parameters ***(PID)*** showed negligible impact, suggesting that the process stability is well-managed, and future optimization efforts should focus on Energy Efficiency and Logistics rather than machine-level tuning.

**Let's also check the equation that dictates this factory**

```python
model = models["Ridge Regression"]
intercepto = model.intercept_

termos = sorted(zip(model.coef_, X.columns), key=lambda x: abs(x[0]), reverse=True)

print ("-> Factory's Equation:")
print("="*30)
print(f"Production_Rate = {intercepto:.2f}")
for peso, nome in termos:
    sinal = "+" if peso >= 0 else "-"
    print(f"   {sinal} ({abs(peso):.2f} * {nome})")
```

**In the previous steps the Ridge Regression model achieved high accuracy (~91%) by including variables like Power Comsuption and Vibration Level. These features are great for monitoring the current state, they represent effects, not causes. Now we want to transition from Monitoring to Controling, so let's retrain the model using only controllable inputs.**

**For a true Digital twin control strategy, we must avoid the reverse causality. We cannot increase production by simply forcing the power consumption down; we must adjust the actual control levers. Therefore, we filter the dataset to include only:**

1.  **Control Levers:** Variables we can physically adjust (e.g., `PID` settings, `Supply_Flow`).
2.  **External Drivers:** Variables imposed by the environment (e.g., `Demand_Response`).

**We expect a drop in the RÂ² score because we are removing the highly correlated symptom variables. However, the resulting model will be physically valid for simulation and decision-making.**

```python
causal_features = [
    'Demand_Response',
    'Supply',
    'PID_Kp','PID_Ki','PID_Kd',
    'Federated_Model_Version',
    'Smart_Contract_Triggered'
]

valid_features = [c for c in causal_features if c in df.columns]

X_causal = df[valid_features]
y_causal = df['Production_Rate']

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_causal,y_causal, test_size = 0.2, random_state = 42)

scaler_c = StandardScaler()
X_train_c = pd.DataFrame(scaler_c.fit_transform(X_train_c), columns = X_train_c.columns)
X_test_c = pd.DataFrame(scaler_c.transform(X_test_c), columns =X_test_c.columns)

causal_model = Ridge(alpha=1.0)
causal_model.fit(X_train_c, y_train_c)

preds_c = causal_model.predict(X_test_c)
r2_c = r2_score(y_test_c, preds_c)
mae_c = mean_absolute_error(y_test_c, preds_c)

print(f"   -> RÂ² Score: {r2_c:.2%}")
print(f"   -> MAE Error: {mae_c:.4f}")

coefs_c = pd.DataFrame({
    'Feature': X_train_c.columns,
    'Weight': causal_model.coef_,
    'Abs_Weight': abs(causal_model.coef_)
}).sort_values(by='Abs_Weight', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Weight', y='Feature', data=coefs_c, palette='RdBu') # Red-Blue diverge
plt.title('True Control Levers: What ACTUALLY controls the factory?', fontsize=14)
plt.xlabel('Impact (Negative = Hinder | Positive = Boost)')
plt.axvline(x=0, color='black', linestyle='--')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

**The RÂ² Score really dropped to ~73%, let's try XGBoost and SVR again because the relations are not linear anymore**

```python
models_c = {
    "Ridge (Linear)": Ridge(alpha = 1.0),
    "SVR": SVR(C=1.0, epsilon = 0.1),
    "XGBoost": xgb.XGBRegressor(objective = 'reg:squarederror', n_estimators = 100, random_state = 42)
}

results_c = {}
for name, model in models_c.items():
    model.fit(X_train_c, y_train_c)
    preds = model.predict(X_test_c)

    r2 = r2_score(y_test_c, preds)
    mae = mean_absolute_error(y_test_c, preds)

    results_c[name] = r2

    print(f"Model: {name}")
    print(f"   -> RÂ² Score: {r2:.2%}")
    print(f"   -> MAE Error: {mae:.4f}")
    print("-" * 30)

    best_model_c = max(results_c, key=results_c.get)
print(f"\n WINNER (Causal Mode): {best_model_c} with RÂ² = {results_c[best_model_c]:.2%}")
```

**Winner:** Ridge Regression (Linear Model)
**Performance:** RÂ² = 73.48 (on Causal Data)

**Key Insights for the Digital Twin:**
1.  **Linearity Confirmed:** The Linear model outperformed the Non-Linear Ensemble (XGBoost: 66.51%), proving that the underlying production dynamics are fundamentally linear. This validates the use of standard **PID Control theory** without needing complex Neural Networks for actuation.
2.  **The "Actuation Gap":** Moving from Monitoring (91% accuracy) to Control (73% accuracy) reveals that ~18% of the variance is driven by unmeasured physical disturbances (friction, thermal drift) that manifest in Power Consumption but are not captured by current input sensors.
3.  **Recommendation:** For the final Digital Twin implementation, we recommend a **Hybrid Approach**:
    * Use **Ridge Regression** for the feed-forward control loop (setting the baseline speed based on Demand).
    * Use the **Power Consumption monitoring** (from the first model) as a feedback mechanism to correct the drift in real-time.

```python
hours_to_simulate = 100
target_production = 150.0
current_production = 120.0

history = {
    'Hour': [],
    'Demand (Disturbance)': [],
    'Control_Action (PID_Kp)': [],
    'Production_Real (Simulated)': [],
    'Target': []
}

current_state = X_test_c.iloc[0].copy()

idx_demand = list(X_test_c.columns).index('Demand_Response')
idx_kp = list(X_test_c.columns).index('PID_Kp')

print(f"   -> Target Setpoint: {target_production} units/h")
print(f"   -> Simulation Horizon: {hours_to_simulate} hours")

for h in range(hours_to_simulate):

    market_demand = np.random.normal(0, 1)
    current_state.iloc[idx_demand] = market_demand

    prediction_input = pd.DataFrame([current_state], columns=X_test_c.columns)
    predicted_output = causal_model.predict(prediction_input)[0]

    error = target_production - predicted_output

    control_gain = 0.5
    previous_kp = current_state.iloc[idx_kp]
    new_kp = previous_kp + (error * control_gain)

    new_kp = np.clip(new_kp, -160.0, 160.0)

    current_state.iloc[idx_kp] = new_kp

    final_input = pd.DataFrame([current_state], columns=X_test_c.columns)
    final_production = causal_model.predict(final_input)[0]

    history['Hour'].append(h)
    history['Demand (Disturbance)'].append(market_demand)
    history['Control_Action (PID_Kp)'].append(new_kp)
    history['Production_Real (Simulated)'].append(final_production)
    history['Target'].append(target_production)

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(history['Hour'], history['Target'], 'r--', label='Target Setpoint (150)', linewidth=2)
plt.plot(history['Hour'], history['Production_Real (Simulated)'], 'g-', label='Digital Twin Output', linewidth=2)
plt.title('Digital Twin Performance: Maintaining Target despite Chaos', fontsize=14)
plt.ylabel('Production Rate')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 1, 2)
plt.plot(history['Hour'], history['Demand (Disturbance)'], 'grey', linestyle=':', label='Market Demand (Uncontrollable)', alpha=0.7)
plt.plot(history['Hour'], history['Control_Action (PID_Kp)'], 'b-', label='Controller Action (PID_Kp Adjustment)', linewidth=2)
plt.title('Behind the Scenes: How the Controller reacts to Demand', fontsize=12)
plt.ylabel('Standardized Values (Z-Score)')
plt.xlabel('Time (Hours)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
